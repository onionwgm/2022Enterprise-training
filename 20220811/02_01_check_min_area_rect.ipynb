```python
# 导包
import cv2 as cv
import numpy as np
```

```python
# 初始化设置图片文件位置
plate_file_path = 'images/restoration1.jpg'
```

```python
# 0. 加载图片
plate_image = cv.imread(plate_file_path)
```

```python
# cv.imshow('plate_image', plate_image)
# cv.waitKey()
# cv.destroyAllWindows()
```

```python
# 1. 图片预处理
# 高斯模糊
blured_image = cv.GaussianBlur(plate_image, (5, 5), 0)
# cv.imshow('blured_image', blured_image)
# cv.waitKey()
# cv.destroyAllWindows()
# 转成灰度图
gray_image = cv.cvtColor(blured_image, cv.COLOR_BGR2GRAY)
# cv.imshow('gray_image', gray_image)
# cv.waitKey()
# cv.destroyAllWindows()
# 使用Sobel算子，求水平方向一阶导数
# 使用 cv.CV_16S
grad_x = cv.Sobel(gray_image, cv.CV_16S, 1, 0, ksize=3)
# 转成 CV-_8U - 借助 cv.convertScaleAbs()方法
abs_grad_x = cv.convertScaleAbs(grad_x)
# 叠加水平和垂直（此处不用）方向，获取 sobel 的输出
gray_image = cv.addWeighted(abs_grad_x, 1, 0, 0, 0)
#gray_image = cv.cvtColor(blured_image, cv.COLOR_BGR2GRAY)
# cv.imshow('gray_image', gray_image)
# cv.waitKey()
# cv.destroyAllWindows()
#二值化操作
is_success, threshold_image = cv.threshold(gray_image, 0, 255, cv.THRESH_OTSU)
# cv.imshow('threshold_image', threshold_image)
# cv.waitKey()
# cv.destroyAllWindows()
# 执行闭操作=>车牌连成矩形区域
kernel = cv.getStructuringElement(cv.MORPH_RECT, (10, 3))
morphology_imge = cv.morphologyEx(threshold_image, cv.MORPH_CLOSE, kernel)
# cv.imshow('morphology_imge', morphology_imge)
# cv.waitKey()
# cv.destroyAllWindows()
# 获取等值线（轮廓）
contours, _ = cv.findContours(morphology_imge, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
# 绘制等值线(在原始图片上)
cv.drawContours(plate_image, contours, -1, (0, 0, 255))
# cv.imshow('drawContours', plate_image)
# cv.waitKey()
# cv.destroyAllWindows()
```

```python
# cv.imshow('drawContours', plate_image)
# cv.waitKey()
# cv.destroyAllWindows()
```

```python
# 获取第0个轮廓
contour = contours[0]
# 获取矩形特征描述的等值线区域，返回：中心点坐标、长和宽、旋转角度
rect = cv.minAreaRect(contour)
# 获取矩形的角点
box = cv.boxPoints(rect)
# 转换为整型
box = np.int0(box)
# 遍历矩形的所有角点
for i in np.arange(len(box)):
    # 在矩形的角点上绘制红色圆圈，给出半径和线宽
    # 注意： box[i],是向量形式表示的角点的坐标，需要转换为元组（tuple）类型
    cv.circle(plate_image, tuple(box[i]),5, (0, 0, 255), 3)
```

```python
# cv.imshow('drawContours', plate_image)
# cv.waitKey()
# cv.destroyAllWindows()
```

```python
# boundingRect 用于获取与 等值线框（轮廓框）contour 的四个角点正交的矩形
# 返回 左上的坐标（x, y），宽（w）、高（h）
x, y, w, h = cv.boundingRect(contour)
# 在原始图片上以线宽为2的蓝色线绘制和轮廓正交的矩形
cv.rectangle(plate_image, (x, y), (x+w, y+h), (255, 0, 0), 2)
```
