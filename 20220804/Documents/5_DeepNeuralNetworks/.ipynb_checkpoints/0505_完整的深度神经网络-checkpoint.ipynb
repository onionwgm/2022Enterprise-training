## **演练0505：完整的深度神经网络**

### **案例1：完整的模块化深度神经网络的代码**
在上一个模块化神经网络代码的实现的基础上，应用梯度更新策略、Batch Normalization和Dropout，实现完整功能的神经网络
* 本例代码包含【full_connected_neural_net_all.py】【util_common_layer.py】【util_batch_norm_layer.py】【util_dropout_layer.py】
* 验证正确率可达：0.561，测试正确率可达：0.536。但这可能已经是多层神经网络的极限能力了。如果想进一步提升正确率，需要使用卷积神经网络

```python
''' 使用全功能神经网络进行图像分类 '''

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import util_data
from full_connected_neural_net_all import FullyConnectedNet
from solver import Solver
from optim import *


# 装载cifar数据 ===============================<
print("========== 数据装载和预处理 ==========")
# 装载数据
cifar10_dir = '../dataset/cifar-10'
data_list = util_data.load_and_preprocess_cifar10(cifar10_dir)
X_train = data_list['X_train']
y_train = data_list['y_train']
X_val = data_list['X_val']
y_val = data_list['y_val']
X_test = data_list['X_test']
y_test = data_list['y_test']

# 均值化Normalize，所有数据减去每列的平均值(Train, Val和Test虽然行数不同，但是列数都相同)
# 注意，Val和Test集都应减去Train集中的平均值
mean_image = np.mean(X_train, axis=0)
X_train -= mean_image
X_val -= mean_image
X_test -= mean_image

# 检查数据维度
print('训练数据: ', X_train.shape)
print('验证数据: ', X_val.shape)
print('测试数据: ', X_test.shape)
print("========== 数据装载和预处理完成 ======")
# 装载cifar数据 >===========================>

data = {
    'X_train': data_list['X_train'],
    'y_train': data_list['y_train'],
    'X_val': data_list['X_val'],
    'y_val': data_list['y_val'],
    'X_test': data_list['X_test'],
    'y_test': data_list['y_test'],
}

best_model = None
best_solver = None
best_val_acc = 0.0
dropout_choice = [0]
netstructure_choice = [[100, 100]]
for hidden_dim in netstructure_choice:
    for dropout in dropout_choice:
        model = FullyConnectedNet(hidden_dim, reg=0, weight_scale=5e-2,
                            use_batchnorm=True, dropout=dropout)
        solver = Solver(model, data,
                num_epochs=20, batch_size=100,
                update_rule='adam',
                optim_config={
                    'learning_rate': 5e-3
                    },
                print_every=100,
                lr_decay=0.95,
                verbose=True)
        solver.train()
        
        if solver.best_val_acc > best_val_acc:
            best_val_acc = solver.best_val_acc
            best_model = model
            best_solver = solver
            
plt.subplot(2, 1, 1)
plt.title('Training loss')
plt.plot(best_solver.loss_history, 'o')
plt.xlabel('Iteration')

plt.subplot(2, 1, 2)
plt.title('Accuracy')
plt.plot(best_solver.train_acc_history, '-o', label='train')
plt.plot(best_solver.val_acc_history, '-o', label='val')
plt.plot([0.5] * len(best_solver.val_acc_history), 'k--')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.gcf().set_size_inches(15, 12)
plt.show()

y_test_pred = np.argmax(best_model.loss(data['X_test']), axis=1)
y_val_pred = np.argmax(best_model.loss(data['X_val']), axis=1)
print('最佳验证数据正确率: ', (y_val_pred == data['y_val']).mean())      # 约0.561
print('最佳测试数据正确率: ', (y_test_pred == data['y_test']).mean())    # 约0.536
```
